```python
# Assuming df is your PySpark DataFrame with columns "Datetime" and "Temperature"
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.functions import col
import matplotlib.pyplot as plt
import warnings

# Suppress pandas FutureWarning
warnings.simplefilter(action='ignore', category=FutureWarning)

spark = SparkSession.builder.appName("dataset").getOrCreate()
df = spark.read.csv(r"C:\Users\S554978\Desktop\bigdata\powerconsumption.csv",header=True,inferSchema=True)


# Extract date and temperature columns
data_for_plot = df.select("Datetime", "Humidity").toPandas()

# Plotting the line chart
plt.figure(figsize=(12, 6))
plt.plot(data_for_plot["Datetime"], data_for_plot["Humidity"], marker='o')
plt.title("Examining humidity levels and fluctuations throughout the time period.")
plt.xlabel("Datetime")
plt.ylabel("Humidity")
plt.grid(True)
plt.show()

```
